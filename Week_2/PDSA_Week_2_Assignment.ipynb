```python
def binarySearchIndexAndComparisons(L, k):
  left = 0
  right = len(L) - 1
  middle = 0
  count = 0
  while(left <= right):
    middle = (left + right) // 2
    if(L[middle] == k):
      return True
    elif(k > L[middle]):
      left = middle + 1
    else:
      right = middle
  return False
```

### Approach 2

```python
def binarySearchIndexAndComparisons(L, k):
  def binary_search(L, k, count):
    if L == []:
      return (False, count)

    middle = (len(L) - 1) // 2

    if L[middle] == k:
      count += 1
      return (True, count)
    elif k < L[middle]:
      count += 1
      return binary_search(L[:middle], k, count)
    else:
      count += 1
      return binary_search(L[middle+1:], k, count)
  
  return (binary_search(L, k, 0))
```

## GrPA_1

```python
def merge_lists_1(A, B):
  """Returns Sorted List of elements of A and B"""
  merged_list = []
  i, j, k = 0, 0, 0
  # i -> Pointer for A
  # j -> Pointer for B
  # k -> Pointer for the merged_list
  while k < (len(A) + len(B)): 
    # While there are elements to be merged
    if i == len(A): 
      # When A is empty
      merged_list.extend(B[j:])
      k += len(B) - j
    elif j == len(B):
      # When B is empty
      merged_list.extend(A[i:])
      k += len(A) - i
    
    # Compare top element of pile A with top element of pile B
    elif A[i] < B[j]:
      merged_list.append(A[i])
      i += 1
      k += 1
    else:
      merged_list.append(B[j])
      j += 1
      k += 1
  return merged_list

def merge_sort_1(L):
  """Returns Sorted List L using Merge Sort"""
  if len(L) <= 1:
    return L
  
  A = merge_sort_1(L[:(len(L))//2])
  B = merge_sort_1(L[len(L)//2:])

  merged_list = merge_lists_1(A, B)

  return merged_list

def merge_lists_2(A, B):
  """Returns Sorted List of elements of A and B"""
  merged_list = []
  i, j, k = 0, 0, 0
  # i -> Pointer for A
  # j -> Pointer for B
  # k -> Pointer for the merged_list
  while k < (len(A) + len(B)): 
    # While there are elements to be merged
    if i == len(A): 
      # When A is empty
      merged_list.extend(B[j:])
      k += len(B) - j
    elif j == len(B):
      # When B is empty
      merged_list.extend(A[i:])
      k += len(A) - i
    
    # Compare top element of pile A with top element of pile B
    elif int(A[i][1:]) > int(B[j][1:]):
      merged_list.append(A[i])
      i += 1
      k += 1
    else:
      merged_list.append(B[j])
      j += 1
      k += 1
  return merged_list

def merge_sort_2(L):
  """Returns Sorted List L using Merge Sort"""
  if len(L) <= 1:
    return L
  
  A = merge_sort_2(L[:(len(L))//2])
  B = merge_sort_2(L[len(L)//2:])

  merged_list = merge_lists_2(A, B)

  return merged_list

def combinationSort(l):
  # Sort type 1
  d = {}
  for i in l:
    d[i[0]] = []
  
  for i in l:
    d[i[0]] += [i]

  list_of_keys = merge_sort_1(list(d.keys()))

  L1 = []
  for i in list_of_keys:
    L1 += d[i]
  
  for i in list_of_keys:
    d[i] = merge_sort_2(d[i])
  
  L2 = []
  for i in list_of_keys:
    L2 += d[i]
  
  # Sort Type 2
  return (L1, L2)
```

```python
L = ["d34", "g54", "d12", "b87", "g1", "c65", "g40", "g5", "d77"]

print(combinationSort(L))
```

### Approach_1

```python
def findLargest(L):
  max = L[0]
  for i in range(1, len(L)):
    if L[i] > max:
      max = L[i]
  
  return max
```

### Approach_2

```python
def findLargestRecursive(L, start, end):
  # Base Case
  if start == end:
    return L[start]
  
  mid = start + (end - start) // 2
  # First element is Maximum
  if mid == 0 and L[mid] > L[mid + 1]:
    return L[mid]
  
  # Current element is maximum
  if mid < end and mid > 0 and L[mid] > L[mid + 1] and L[mid] > L[mid - 1]:
    return L[mid]
  
  # If maximum is in left half
  if L[start] > L[mid]:
    return findLargestRecursive(L, start, mid - 1)
  
  # If maximum is in right half
  else:
    return findLargestRecursive(L, mid + 1, end)
```

```python
def findLargest(L):
  return findLargestRecursive(L, 0, len(L) - 1)
```

```python
L = [7,8,2,4,5,6]
print(findLargest(L))
```

## GrPA_3

```python
def mergeInPlace(A, B):
  for i in range(len(A)):
    for j in range(len(B)):
      if A[i] > B[j]:
        # Swap A[i] and B[j]
        A[i], B[j] = B[j], A[i]
  
  for i in range(len(B)):
    for j in range(i+1, len(B)):
      if B[i] > B[j]:
        # Swap A[i] and B[j]
        B[i], B[j] = B[j], B[i]
```

```python
A = [2, 4, 6, 9, 13, 15]
B = [1, 3, 5, 10]
mergeInPlace(A, B)
print(A)
print(B)

A = [4, 6]
B = [1, 3, 6, 10]
mergeInPlace(A, B)
print(A)
print(B)
```
