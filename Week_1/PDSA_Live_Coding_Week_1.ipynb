## Quesiton_1

```python
def is_prime(n):
  if n == 2 or n == 3:
    return True
  if n%2 == 0 or n < 2:
    return False
  for i in range(3, int(n**0.5)+1, 2):
    if n%i == 0:
      return False
  
  return True
```

```python
def prime_product(n):
  if n <= 2:
    return False
  for p in range(2, n):
    if n%p == 0:
      if is_prime(p) and is_prime(n // p):
        return True
      else:
        return False
```

## Question_2

```python
def del_char(s, c):
  if (len(c) != 1):
    return s
  return s.replace(c, "")
```

```python
s = "banana"
c = "b"

print(del_char(s, c))

c = "an"
print(del_char(s, c))
```

## Question_3

```python
def shuffle(l1, l2):
  l = []
  if len(l1) == len(l2):
    for i in range(len(l1)):
      l.append(l1[i])
      l.append(l2[i])
  elif len(l1) > len(l2):
    i = 0
    while(i < len(l2)):
      l.append(l1[i])
      l.append(l2[i])
      i += 1
    
    # l1 will still have some elements
    while(i < len(l1)):
      l.append(l1[i])
      i += 1
  else:
    i = 0
    while(i < len(l1)):
      l.append(l1[i])
      l.append(l2[i])
      i += 1
    
    # l1 will still have some elements
    while(i < len(l2)):
      l.append(l2[i])
      i += 1
  
  return l
```

```python
L1 = [0,2,4]
L2 = [1,3,5]

print(shuffle(L1,L2))

L1 = [0,2,4]
L2 = [1]

print(shuffle(L1,L2))

L1 = [0]
L2 = [1,3,5]

print(shuffle(L1,L2))
```

## Question_4

```python
def expanding(L):
  if len(L) < 2:
    return True
  diff = abs(L[1] - L[0])
  for i in range(1, len(L)-1):
    if(abs(L[i+1] - L[i]) <= diff):
      return False
    else:
      diff = abs(L[i+1] - L[i])
  return True
```

```python
L = [1,3,7,2,9]
print(expanding(L))

L = [1,3,7,2,-3]
print(expanding(L))

L = [1,3,7,10]
print(expanding(L))
```

## Question_5

```python
def sumsquare(l):
  sum_of_squares = [0, 0]
  for i in l:
    if i%2==0:
      sum_of_squares[1] += i**2
    else:
      sum_of_squares[0] += i**2
  return sum_of_squares
```

```python
L = [1,3,5]
print(sumsquare(L))

L = [-1,-2,3,7]
print(sumsquare(L))

L = [2,4,6]
print(sumsquare(L))
```

## Question_6

```python
def histogram(l):
  d = {}
  final_list = []

  for i in set(l):
    d[i] = 0
  
  for i in l:
    d[i] += 1
  
  for i in d:
    final_list.append((i, d[i]))
  
  # Sort by keys or values
  for i in range(len(final_list)):
    for j in range(i+1, len(final_list)):
      if final_list[i][0] > final_list[j][0] or final_list[i][1] > final_list[j][1]:
        final_list[i], final_list[j] = final_list[j], final_list[i]
  
  return final_list
  
```

```python
print(histogram([7,12,11,13,7,11,13,14,12]))
print(histogram([13,7,12,7,11,13,14,13,7,11,13,14,12,14,14,7]
))
print(histogram([13,12,11,13,14,13,7,7,13,14,12]))
```
